# Handling Missing Values

```python
# filling the missing value of ENGINE HP with MEAN of all values
```

```python
mean_hp = data["Engine HP"].mean()
data["Engine HP"].fillna(mean_hp, inplace=True)
```

```python
# filling the missing value of ENGINE CYLINDERS with MEAN of all values
```

```python
mean_Cylinders = data["Engine Cylinders"].mean()
data["Engine Cylinders"].fillna(mean_Cylinders, inplace=True)
```

```python
data.isna().sum()
```

```python
# filling the missing value of ENGINE FUEL TYPE with MODE of all values
```

```python
mode_fuel_type = data["Engine Fuel Type"].mode()[0]
data["Engine Fuel Type"].fillna(mode_fuel_type, inplace=True)
```

```python
# filling the missing value of NUMBER OF DOORS with MODE of all values
```

```python
mode_Doors = data["Number of Doors"].mode()[0]
data["Number of Doors"].fillna(mode_Doors, inplace=True)
```

```python
data.isna().sum()
```

```python
# filling the missing value of MARKET CATEGORY as NORMAL as there is a large data so we cannot afford to drop it 
```

```python
data["Market Category"].fillna("NORMAL", inplace=True)
```
